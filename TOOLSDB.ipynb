```python
from openpyxl import load_workbook
from openpyxl.utils import get_column_letter
from copy import copy
import random
import math
# Load the existing workbook
# \u278A = 1, \u278B = 2, \u278C = 3, \u278D = 4
workbook = load_workbook(filename="hello1.xlsx")
source_sheet = workbook['Default']
#xet tong cong 8 lop
for i in range(len(excel_dict)):
    #o lop thu i, kiem tra xem can tao bao nhieu sheet bang cach lay chieu dai cua thu chia 8, vi moi sheet co 8 row.
    for j in range(math.ceil(len(excel_dict[i][3])/8)):
        mamon, malop = excel_dict[i][0][0].rsplit(' ', 1)
        sheetname = mamon+" "+malop+str(j+1)
        if sheetname in workbook.sheetnames:
            print(f"Sheet {sheetname} already exists. Handling accordingly.")
            # Optional: Remove existing sheet or choose a different name
        else:
            target_sheet = workbook.create_sheet(sheetname)
        # for row in source.rows:
        #     for cell in row:
        #         target[cell.coordinate].value = cell.value
        for row in source_sheet.iter_rows():
                for cell in row:
                    new_cell = target_sheet.cell(row=cell.row, column=cell.column, value=cell.value)
                    # Copy cell style if needed
                    if cell.has_style:
                        new_cell.font = copy(cell.font)
                        new_cell.border = copy(cell.border)
                        new_cell.fill = copy(cell.fill)
                        new_cell.number_format = copy(cell.number_format)
                        new_cell.protection = copy(cell.protection)
                        new_cell.alignment = copy(cell.alignment)
                # Copy the row height from source to target
                target_sheet.row_dimensions[cell.row].height = source_sheet.row_dimensions[cell.row].height

            # Copy column widths
        for col in source_sheet.columns:
            # Using the first cell to determine the column letter
            column_letter = get_column_letter(col[0].column)
            target_sheet.column_dimensions[column_letter].width = source_sheet.column_dimensions[column_letter].width

        worksheet = workbook[sheetname]
            #worksheet["H3"] = "\u278A"
        worksheet["H3"] = excel_dict[i][1][0]
        mamon, malop = excel_dict[i][0][0].rsplit(' ', 1)
        worksheet["L3"] = mamon
        worksheet["M3"] = "Lớp: "+malop
        worksheet["H4"] = 'Phạm Minh Vân Anh'
        worksheet["M1"] = 'Trang số: '+sheetname[-1]
        day_to_number = {
        'Monday': 2,
        'Tuesday': 3,
        'Wednesday': 4,
        'Thursday': 5,
        'Friday': 6,
        'Saturday': 7,
        'Sunday': "CN"
        }
        #Insert thu
        start_index = j * 8  # Starting index for slicing excel_dict[i][4]
        end_index = start_index + 8  # End index for slicing, ensuring not to exceed list bounds
        for k, day in enumerate(excel_dict[i][4][start_index:end_index], start=11):
            worksheet[f"A{k}"] = day_to_number[day]
        #Insert ngay thang nam
        for k, date in enumerate(excel_dict[i][3][start_index:end_index], start=11):
            worksheet[f"B{k}"] = date
        #Insert buoi giang
        day_to_scharacter = {
        'Sang': '\u25A0'+'\u25A1'+'\u25A1',
        'Chieu': '\u25A1'+'\u25A0'+'\u25A1',
        'Toi': '\u25A1'+'\u25A1'+'\u25A0'
        }
        for k, day in enumerate(excel_dict[i][7][start_index:end_index], start=11):
            worksheet[f"D{k}"] = day_to_scharacter[day]
        #Insert thu tu gio
        hour_to_scharacter = {
        # \u278A = 1, \u278B = 2, \u278C = 3, \u278D = 4
        '1': '\u278A',
        '2': '\u278B',
        '3': '\u278C',
        '4': '\u278D'
        }
        for k, hour in enumerate(excel_dict[i][6][start_index:end_index], start=11):
            finalscharacter=''
            split1234 = [char for char in hour]
            for num in split1234:
                finalscharacter=finalscharacter+' '+hour_to_scharacter[num]
            worksheet[f"E{k}"] = finalscharacter
        #Insert tong so gio
        for k, totalhour in enumerate(excel_dict[i][8][start_index:end_index], start=11):
            worksheet[f"F{k}"] = totalhour
        #Insert phong hoc
        for k, room in enumerate(excel_dict[i][9][start_index:end_index], start=11):
            worksheet[f"G{k}"] = room
        #Insert co so
        for k, venue in enumerate(excel_dict[i][10][start_index:end_index], start=11):
            worksheet[f"H{k}"] = venue
        #Insert si so sv
        for k, venue in enumerate(excel_dict[i][10][start_index:end_index], start=11):
            worksheet[f"I{k}"] = random.randint(28, 32)
        #Insert noi dung
        for k, lesson_content in enumerate(excel_dict[i][5][start_index:end_index], start=11):
            worksheet[f"J{k}"] = lesson_content

# # Select a worksheet by its name
# worksheet_name = "Sheet2"  # Replace "SheetName" with the actual name of your sheet
# if worksheet_name in workbook.sheetnames:
#     worksheet = workbook[worksheet_name]

    #worksheet.add_image(Image("so4.png"), "O9")
    # Save the workbook to keep changes
workbook.save(filename="SDB_VanAnh.xlsx")
# else:
#     print(f"Worksheet {worksheet_name} does not exist in the workbook.")
```
